```python
import os
import time
import pandas as pd
import warnings

warnings.filterwarnings("ignore")
```

```python
root_path = os.path.abspath(".")
op_dir_name = "新股网下数据周报"
op_file_name = "新股网下申购配售统计"
op_file_type = ".xlsx"
sh_tail = 1
sz_tail = 2
```

```python
dir_list = os.listdir(root_path)
```

```python
data_list = [item for item in os.listdir(root_path) if op_dir_name in item]
data_list
```

```python
file_date = data_list[-1].split(op_dir_name)[-1]
op_date = time.strftime('%Y%m%d',time.localtime(time.time()))
```

```python
if file_date <= op_date:
    file_path = os.path.join(root_path, op_dir_name + file_date, op_file_name + file_date + op_file_type)
    print("Deal with the {} data, path is: {}".format(file_date, file_path))
```

```python
df = pd.read_excel(file_path, header=[2, 3], sheet_name=["上海", "深圳"])
df_sz = df["深圳"].tail(sz_tail)
df_sz
```

```python
col_name = [item[0] for item in df_sz.keys()]
col_name
```

```python
del_list = [
    '申购日',
    '发行价',
    '个人金额',
    '产品金额',
    '备注',
    '个人——获配数量（股）',
    '个人——获配数量（股）',
    '个人——获配数量（股）',
    '个人——获配数量（股）',
    '个人——获配数量（股）',
    '个人——获配数量（股）',
    '个人——获配数量（股）',
    '个人——获配数量（股）',
    '明湾',
    '明湾',
    '金伙伴',
    '民盛系列',
    '民盛系列',
    '潜龙系列',
    '潜龙系列',
    '潜龙系列',
    '潜龙系列',
]
```

```python
drop_list = list(set(del_list))
```

```python
info_list = ['代码', '名称']
```

```python
df_sz_new = df_sz.drop(drop_list, axis=1)
```

```python
df_sz_new
```

```python
temp_keys = df_sz_new.keys()
temp_keys
```

```python
product_list = list()
value_list = list()
for item in temp_keys:
    value_list.append(df_sz_new[item].tolist())
    if 'Unnamed:' not in item[1]:
        product_list.append(''.join(item).replace('龙凤呈祥', '龙凤'))
    else:
        product_list.append(item[0])
    len(product_list)

# product_list = list(set(product_list) - set(info_list))
```

```python
ns_sz = dict()
for key, value in zip(product_list, value_list):
    ns_sz[key] = value
    
ns_sz
```

```python
pb_list = pd.read_excel("pb_list.xlsx", header=None, index_col=None)[0].tolist()
```

```python
pb_list
```

```python
zzz = pd.read_excel("zzz.xlsx", header=0, index_col="账户")
```

```python
zzz
```

```python
zzz.loc["和谐1号"]
```

```python
ns_col = ["序号", "账户", "账号", "券商", "交易员"]
ns_col_sz = ns_col.copy()

for name, idx in zip(ns_sz["名称"], ns_sz["代码"]):
    ns_col_sz.append(name + str(idx).rjust(6, '0'))

ns_col_sz
```

```python
ns_op_sz = pd.DataFrame(columns=ns_col_sz)
```

```python
ns_op_sz
```

```python
for item in info_list:
    if item in ns_sz.keys():
        ns_sz.pop(item)
```

```python
idx = 0

for key, value in ns_sz.items():
    idx += 1

    if key in zzz.index:
        zh_item = str(zzz.loc[key][ns_col_sz[2]])
        qs_item = zzz.loc[key][ns_col_sz[3]]
    else:
        zh_item = ""
        qs_item = ""

    item = {
        ns_col_sz[0]: idx,
        ns_col_sz[1]: key,
        ns_col_sz[2]: zh_item,
        ns_col_sz[3]: qs_item,
        ns_col_sz[4]: "operator"
    }

    for v_idx in range(len(value)):
        item[ns_col_sz[5 + v_idx]] = value[v_idx]
    
    ns_op_sz = ns_op_sz.append(item, ignore_index=True)
    
ns_op_sz
```

```python
judge = [False] * len(ns_op_sz)
# print(judge)

for idx in range(sz_tail):
#     print(ns_op_sz[ns_col_sz[- 1 - idx]].notna().tolist())
    judge = judge or ns_op_sz[ns_col_sz[- 1 - idx]].notna().tolist()
```

```python
ns_op_sz.to_excel("新股周报.xlsx", sheet_name="深圳", index=None)
```

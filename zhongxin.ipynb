```python
for sheet in zcz.sheet_names:
    df  = pd.read_excel("注册制.xlsx", sheet_name=sheet, header=1)
    for item in zx_list:
        je = 0
        sl = 0
        if item in df["证券账户"].tolist():
            idx = df[df["证券账户"] == item].index[0]
            je += df["金额"][idx]
            sl += df["数量"][idx]
    hss["中信产品"][sheet] = je / sl
    hss["其他"][sheet] = (df["金额"][2] - je) / (df["数量"][2] - sl)
```

```python
hss.to_excel("均价对比.xlsx")
```
